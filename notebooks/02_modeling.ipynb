# 02 - Machine Learning Modeling & Threshold Tuning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')
from src.train_model import train_and_evaluate
from src.evaluate_model import evaluate_thresholds
```

## 1. Train Models

```python
results_df, best_model, X_test, y_test = train_and_evaluate()
display(results_df)
```

## 2. Threshold Tuning

```python
thresh_df = evaluate_thresholds(best_model, X_test, y_test, thresholds=[0.3, 0.4, 0.5, 0.6, 0.7])
display(thresh_df)
```

## 3. Visualization of Tradeoffs

```python
plt.figure(figsize=(10,5))
plt.plot(thresh_df['Threshold'], thresh_df['Precision'], marker='o', label='Precision')
plt.plot(thresh_df['Threshold'], thresh_df['Recall/Sensitivity'], marker='o', label='Recall')
plt.legend()
plt.title('Precision vs Recall at Different Thresholds')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.show()
```
