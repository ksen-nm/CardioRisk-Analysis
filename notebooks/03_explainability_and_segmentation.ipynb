# 03 - Explainability & Risk Segmentation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import joblib
sys.path.append('..')
from src.explain_model import get_feature_importance, get_permutation_importance
from src.train_model import prepare_data
```

## 1. Feature Importance

```python
model = joblib.load('../models/best_model.pkl')
X_train, X_test, y_train, y_test = prepare_data('../data/processed/cardio_features.csv')

feat_imp = get_feature_importance(model, X_train)
if feat_imp is not None:
    plt.figure(figsize=(10,6))
    sns.barplot(data=feat_imp, x='Importance', y='Feature')
    plt.title('Feature Importance')
    plt.show()
```

## 2. Permutation Importance

```python
perm_imp = get_permutation_importance(model, X_test, y_test)
plt.figure(figsize=(10,6))
sns.barplot(data=perm_imp, x='Importance', y='Feature')
plt.title('Permutation Importance')
plt.show()
```
