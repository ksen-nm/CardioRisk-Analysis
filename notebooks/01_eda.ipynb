# 01 - Exploratory Data Analysis & Statistical Analysis

## 1. Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, ttest_ind

df = pd.read_csv('../data/processed/cardio_features.csv')
df.head()
```

## 2. Basic Info & Missing Values

```python
print(df.info())
print(df.isnull().sum())
print('Duplicates:', df.duplicated().sum())
```

## 3. Visualizations

```python
plt.figure(figsize=(6,4))
sns.countplot(x='cardio', data=df)
plt.title('Target Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='age_years', hue='cardio', kde=True)
plt.title('Age Distribution by CVD')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='BMI', hue='cardio', kde=True, bins=50)
plt.xlim(10, 50)
plt.title('BMI Distribution by CVD')
plt.show()
```

## 4. Statistical Analysis

```python
def run_chi2(col):
    contingency = pd.crosstab(df[col], df['cardio'])
    c, p, dof, expected = chi2_contingency(contingency)
    print(f'{col} - Chi2 p-value: {p:.4e}')

for col in ['cholesterol', 'gluc', 'gender', 'smoke', 'alco', 'active', 'bmi_category', 'age_group', 'bp_category']:
    if col in df.columns:
        run_chi2(col)
```

```python
def run_ttest(col):
    g0 = df[df['cardio']==0][col]
    g1 = df[df['cardio']==1][col]
    t, p = ttest_ind(g0, g1, equal_var=False)
    print(f'{col} - T-test p-value: {p:.4e}')

for col in ['age_years', 'BMI', 'ap_hi', 'ap_lo', 'pulse_pressure']:
    if col in df.columns:
        run_ttest(col)
```
